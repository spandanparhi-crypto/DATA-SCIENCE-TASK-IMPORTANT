```python
# 5. Count total characters in a file

with open("sample.txt", "w") as file:
    file.write("Hello Python")

with open("sample.txt", "r") as file:
    data = file.read()

print("Characters =", len(data))
```

```python
# 8. Read file line by line with line numbers

with open("lang.txt", "r") as file:
    for i, line in enumerate(file, start=1):
        print(i, line.strip())
```

```python
# 9. Copy contents from source file to destination file

with open("source.txt", "w") as file:
    file.write("Hello World")

with open("source.txt", "r") as src:
    data = src.read()

with open("destination.txt", "w") as dest:
    dest.write(data)

print("File copied successfully")
```

```python
# 10. Count occurrences of a specific word

with open("data.txt", "w") as file:
    file.write("Python Java Python C++ Python")

word = input("Enter word: ")

with open("data.txt", "r") as file:
    data = file.read()

count = data.split().count(word)

print("Count =", count)
```

```python
# 11. Store 5 student names using writelines()

students = ["Amit\n", "Rahul\n", "Priya\n", "Neha\n", "Kiran\n"]

with open("students.txt", "w") as file:
    file.writelines(students)

print("Data saved")
```

```python
# 12. Display only lines containing "Python"

with open("topics.txt", "w") as file:
    file.write("Python Basics\n")
    file.write("Java Basics\n")
    file.write("Python Advanced\n")

with open("topics.txt", "r") as file:
    for line in file:
        if "Python" in line:
            print(line.strip())
```

```python
# 13. Store numbers 1 to 100 in a file

with open("numbers.txt", "w") as file:
    for i in range(1, 101):
        file.write(str(i) + "\n")

print("Numbers saved")
```

```python
# 14. Read numbers and calculate sum

with open("nums.txt", "w") as file:
    file.write("10\n20\n30\n40\n")

total = 0

with open("nums.txt", "r") as file:
    for line in file:
        total += int(line.strip())

print("Sum =", total)
```

```python
# 15. Student Management System

while True:
    print("\n1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        name = input("Enter student name: ")
        with open("students.txt", "a") as file:
            file.write(name + "\n")
        print("Student Added")

    elif choice == "2":
        try:
            with open("students.txt", "r") as file:
                print("\nStudents List:")
                print(file.read())
        except FileNotFoundError:
            print("No student records found")

    elif choice == "3":
        name = input("Enter name to search: ")

        try:
            with open("students.txt", "r") as file:
                students = file.read().splitlines()

            if name in students:
                print("Student Found")
            else:
                print("Student Not Found")

        except FileNotFoundError:
            print("No student records found")

    elif choice == "4":
        print("Exiting...")
        break

    else:
        print("Invalid Choice")
```
