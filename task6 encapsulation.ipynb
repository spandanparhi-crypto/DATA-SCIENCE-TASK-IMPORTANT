```python
# Library Book

class Book:
    def __init__(self, book_name, copies):
        self.__book_name = book_name
        self.__copies = copies if copies >= 0 else 0

    def issue_book(self):
        if self.__copies > 0:
            self.__copies -= 1

    def return_book(self):
        self.__copies += 1

    def display(self):
        print("Book:", self.__book_name)
        print("Copies:", self.__copies)


b1 = Book("Python Basics", 5)
b1.issue_book()
b1.return_book()
b1.display()
```

```python
# Bank Account System

class BankAccount:
    def __init__(self, holder, balance):
        self.__account_holder = holder
        self.__balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount

    def check_balance(self):
        print("Balance:", self.__balance)


acc = BankAccount("Spandan", 10000)
acc.deposit(2000)
acc.withdraw(3000)
acc.check_balance()

```

```python
# Student Result Management

class Student:
    def __init__(self, name, python, sql, statistics):
        self.__name = name
        self.__python = python
        self.__sql = sql
        self.__statistics = statistics

    def calculate_total(self):
        return self.__python + self.__sql + self.__statistics

    def calculate_percentage(self):
        return self.calculate_total() / 3

    def display_result(self):
        print("Total:", self.calculate_total())
        print("Percentage:", self.calculate_percentage())


s1 = Student("Spandan", 90, 85, 80)
s1.display_result()
```

```python
# E-Commerce Product

class Product:
    def __init__(self, name, price, stock):
        self.__product_name = name
        self.__price = price
        self.__stock = stock

    def purchase(self, qty):
        if qty <= self.__stock:
            self.__stock -= qty

    def display_product(self):
        print("Product:", self.__product_name)
        print("Stock:", self.__stock)


p1 = Product("Laptop", 50000, 10)
p1.purchase(2)
p1.display_product()
```

```python
class Patient:
    def __init__(self, name, age, bill):
        self.__patient_name = name
        self.__age = age
        self.__bill = bill

    # Getter methods
    def get_name(self):
        return self.__patient_name

    def get_age(self):
        return self.__age

    def get_bill(self):
        return self.__bill

    # Update bill
    def update_bill(self, new_bill):
        if new_bill >= 0:
            self.__bill = new_bill
        else:
            print("Bill cannot be negative")

    # Display details
    def display_patient(self):
        print("Name:", self.__patient_name)
        print("Age:", self.__age)
        print("Bill:", self.__bill)


# Object creation
p1 = Patient("Rahul", 25, 5000)

# Get private variables
print("Name:", p1.get_name())
print("Age:", p1.get_age())
print("Bill:", p1.get_bill())

# Update bill
p1.update_bill(7000)

# Display updated details
p1.display_patient()
```

```python
# Employee Payroll System

class Employee:
    def __init__(self, name, salary):
        self.__employee_name = name
        self.__basic_salary = salary
        self.__bonus = 0

    def add_bonus(self, bonus):
        if bonus >= 0:
            self.__bonus = bonus

    def calculate_salary(self):
        return self.__basic_salary + self.__bonus

    def display_salary(self):
        print("Total Salary:", self.calculate_salary())


e1 = Employee("Spandan", 30000)
e1.add_bonus(5000)
e1.display_salary()
```

```python
# ATM Machine

class ATM:
    def __init__(self, pin, balance):
        self.__pin = pin
        self.__balance = balance

    def withdraw(self, pin, amount):
        if pin == self.__pin and amount <= self.__balance:
            self.__balance -= amount

    def check_balance(self, pin):
        if pin == self.__pin:
            print("Balance:", self.__balance)


atm = ATM(1234, 10000)
atm.withdraw(1234, 2000)
atm.check_balance(1234)
```

```python
# Netflix Account

class Netflix:
    def __init__(self, email, password, plan):
        self.__email = email
        self.__password = password
        self.__plan = plan

    def change_password(self, new_password):
        if len(new_password) >= 8:
            self.__password = new_password

    def upgrade_plan(self, plan):
        self.__plan = plan

    def display_account(self):
        print("Email:", self.__email)
        print("Plan:", self.__plan)


n1 = Netflix("spandan@gmail.com", "password123", "Basic")
n1.upgrade_plan("Premium")
n1.display_account()
```

```python
# Laptop Store

class Laptop:
    def __init__(self, brand, ram, price, stock):
        self.__brand = brand
        self.__ram = ram
        self.__price = price
        self.__stock = stock

    def sell_laptop(self):
        if self.__stock > 0:
            self.__stock -= 1

    def display_details(self):
        print("Brand:", self.__brand)
        print("Stock:", self.__stock)


l1 = Laptop("HP", 8, 50000, 10)
l1.sell_laptop()
l1.display_details()
```
