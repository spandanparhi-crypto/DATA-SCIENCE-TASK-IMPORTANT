```python
# 1. Father + Mother → Child

class Father:
    def __init__(self, father_name, father_age):
        self.father_name = father_name
        self.father_age = father_age

class Mother:
    def __init__(self, mother_name, mother_age):
        self.mother_name = mother_name
        self.mother_age = mother_age

class Child(Father, Mother):
    def __init__(self, father_name, father_age, mother_name, mother_age, child_name):
        Father.__init__(self, father_name, father_age)
        Mother.__init__(self, mother_name, mother_age)
        self.child_name = child_name

    def display(self):
        print("Father Name:", self.father_name)
        print("Father Age:", self.father_age)
        print("Mother Name:", self.mother_name)
        print("Mother Age:", self.mother_age)
        print("Child Name:", self.child_name)

c = Child("Ramesh", 50, "Sita", 45, "Rahul")
c.display()
```

```python
#  Teacher + Engineer + Student

class Teacher:
    def __init__(self, teaching_subject):
        self.teaching_subject = teaching_subject

class Engineer:
    def __init__(self, programming_language):
        self.programming_language = programming_language

class Student(Teacher, Engineer):
    def __init__(self, teaching_subject, programming_language, student_name):
        Teacher.__init__(self, teaching_subject)
        Engineer.__init__(self, programming_language)
        self.student_name = student_name

    def display(self):
        print("Student Name:", self.student_name)
        print("Teaching Subject:", self.teaching_subject)
        print("Programming Language:", self.programming_language)

s = Student("Python", "Java", "Spandan")
s.display()
```

```python
# Mobile + Laptop → Electronics

class Mobile:
    def __init__(self, mobile_brand, **kwargs):
        self.mobile_brand = mobile_brand
        super().__init__(**kwargs)

class Laptop:
    def __init__(self, laptop_brand, **kwargs):
        self.laptop_brand = laptop_brand
        super().__init__(**kwargs)

class Electronics(Mobile, Laptop):
    def __init__(self, mobile_brand, laptop_brand, owner_name):
        super().__init__(
            mobile_brand=mobile_brand,
            laptop_brand=laptop_brand
        )
        self.owner_name = owner_name

    def display(self):
        print("Mobile Brand:", self.mobile_brand)
        print("Laptop Brand:", self.laptop_brand)
        print("Owner Name:", self.owner_name)

e = Electronics("Samsung", "Dell", "Spandan")
e.display()
```

```python
# CricketPlayer + FootballPlayer → Athlete

class CricketPlayer:
    def __init__(self, cricket_team):
        self.cricket_team = cricket_team

class FootballPlayer:
    def __init__(self, football_team):
        self.football_team = football_team

class Athlete(CricketPlayer, FootballPlayer):
    def __init__(self, cricket_team, football_team, player_name):
        CricketPlayer.__init__(self, cricket_team)
        FootballPlayer.__init__(self, football_team)
        self.player_name = player_name

    def display(self):
        print("Player Name:", self.player_name)
        print("Cricket Team:", self.cricket_team)
        print("Football Team:", self.football_team)

a = Athlete("India", "FC Goa", "Virat")
a.display()
```

```python
# Bank + Insurance → Customer (Abstraction + Multiple Inheritance)

from abc import ABC, abstractmethod

class Bank(ABC):
    def __init__(self, bank_name):
        self.bank_name = bank_name

    @abstractmethod
    def show_bank_details(self):
        pass

class Insurance:
    def __init__(self, policy_name):
        self.policy_name = policy_name

class Customer(Bank, Insurance):
    def __init__(self, bank_name, policy_name, customer_name):
        Bank.__init__(self, bank_name)
        Insurance.__init__(self, policy_name)
        self.customer_name = customer_name

    def show_bank_details(self):
        print("Bank Name:", self.bank_name)

    def display(self):
        print("Customer Name:", self.customer_name)
        self.show_bank_details()
        print("Policy Name:", self.policy_name)

c = Customer("SBI", "Life Shield", "Spandan")
c.display()
```

```python
# Animal + Pet → Dog

from abc import ABC, abstractmethod

class Animal(ABC):
    def __init__(self, animal_name):
        self.animal_name = animal_name

    @abstractmethod
    def make_sound(self):
        pass

class Pet:
    def __init__(self, owner_name):
        self.owner_name = owner_name

class Dog(Animal, Pet):
    def __init__(self, animal_name, owner_name):
        Animal.__init__(self, animal_name)
        Pet.__init__(self, owner_name)

    def make_sound(self):
        print("Sound: Bark Bark")

    def display(self):
        print("Animal Name:", self.animal_name)
        print("Owner Name:", self.owner_name)
        self.make_sound()

d = Dog("German Shepherd", "Spandan")
d.display()
```

```python
# Employee + Department → Manager

from abc import ABC, abstractmethod

class Employee(ABC):
    def __init__(self, employee_name):
        self.employee_name = employee_name

    @abstractmethod
    def work(self):
        pass

class Department:
    def __init__(self, department_name):
        self.department_name = department_name

class Manager(Employee, Department):
    def __init__(self, employee_name, department_name):
        Employee.__init__(self, employee_name)
        Department.__init__(self, department_name)

    def work(self):
        print("Work: Managing the department")

    def display(self):
        print("Employee Name:", self.employee_name)
        print("Department Name:", self.department_name)
        self.work()

m = Manager("Spandan", "Data Science")
m.display()
```

```python
# Course + Institute → Student

from abc import ABC, abstractmethod

class Course(ABC):
    def __init__(self, course_name):
        self.course_name = course_name

    @abstractmethod
    def duration(self):
        pass

class Institute:
    def __init__(self, institute_name):
        self.institute_name = institute_name

class Student(Course, Institute):
    def __init__(self, course_name, institute_name, student_name):
        Course.__init__(self, course_name)
        Institute.__init__(self, institute_name)
        self.student_name = student_name

    def duration(self):
        print("Duration: 6 Months")

    def display(self):
        print("Student Name:", self.student_name)
        print("Course Name:", self.course_name)
        print("Institute Name:", self.institute_name)
        self.duration()

s = Student("Full Stack Data Science with GenAI",
            "Naresh IT",
            "Spandan")
s.display()
```

```python
# Vehicle + FuelSystem → Car

from abc import ABC, abstractmethod

class Vehicle(ABC):
    def __init__(self, brand):
        self.brand = brand

    @abstractmethod
    def start(self):
        pass

class FuelSystem:
    def __init__(self, fuel_type):
        self.fuel_type = fuel_type

class Car(Vehicle, FuelSystem):
    def __init__(self, brand, fuel_type, model):
        Vehicle.__init__(self, brand)
        FuelSystem.__init__(self, fuel_type)
        self.model = model

    def start(self):
        print("Car Started Successfully")

    def display(self):
        print("Brand:", self.brand)
        print("Model:", self.model)
        print("Fuel Type:", self.fuel_type)
        self.start()

c = Car("Hyundai", "Petrol", "Creta")
c.display()
```
