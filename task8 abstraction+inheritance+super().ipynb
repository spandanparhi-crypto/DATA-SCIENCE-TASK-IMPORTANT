```python
# Vehicle System

from abc import ABC, abstractmethod

# Parent Abstract Class
class Vehicle(ABC):

    def __init__(self, brand):
        self.brand = brand

    @abstractmethod
    def start(self):
        pass


# Child Class
class Car(Vehicle):

    def __init__(self, brand):
        super().__init__(brand)

    def start(self):
        print(f"{self.brand} car starts with a key")


# Create Object
car1 = Car("Toyota")

# Call Method
car1.start()
```

```python
# Employee Management

from abc import ABC, abstractmethod

class Employee(ABC):

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    @abstractmethod
    def work(self):
        pass


class Developer(Employee):

    def __init__(self, name, salary):
        super().__init__(name, salary)

    def work(self):
        print("Developer writes code")

    def show(self):
        print("Name :", self.name)
        print("Salary :", self.salary)
        self.work()


emp = Developer("Spandan", 50000)
emp.show()
```

```python
# Bank Account

from abc import ABC, abstractmethod

class Bank(ABC):

    def __init__(self, account_holder, account_number):
        self.account_holder = account_holder
        self.account_number = account_number

    @abstractmethod
    def show_interest_rate(self):
        pass


class SBI(Bank):

    def __init__(self, account_holder, account_number):
        super().__init__(account_holder, account_number)

    def show_interest_rate(self):
        print("SBI Interest Rate : 6.5%")


obj = SBI("Spandan", 123456)
print(obj.account_holder)
print(obj.account_number)
obj.show_interest_rate()
```

```python
# 4. Animal Sound System

from abc import ABC, abstractmethod

class Animal(ABC):

    def __init__(self, animal_name):
        self.animal_name = animal_name

    @abstractmethod
    def sound(self):
        pass


class Dog(Animal):

    def __init__(self, animal_name):
        super().__init__(animal_name)

    def sound(self):
        print("Dog says Woof Woof")

    def show(self):
        print("Animal Name :", self.animal_name)
        self.sound()


d = Dog("German Shepherd")
d.show()
```

```python
# 5. Shape Area Calculator

from abc import ABC, abstractmethod

class Shape(ABC):

    def __init__(self, color):
        self.color = color

    @abstractmethod
    def area(self):
        pass


class Rectangle(Shape):

    def __init__(self, color, length, width):
        super().__init__(color)
        self.length = length
        self.width = width

    def area(self):
        print("Area :", self.length * self.width)

    def show(self):
        print("Color :", self.color)
        self.area()


r = Rectangle("Blue", 10, 5)
r.show()
```

```python
# 6. Online Payment System

from abc import ABC, abstractmethod

class Payment(ABC):

    def __init__(self, user_name, amount):
        self.user_name = user_name
        self.amount = amount

    @abstractmethod
    def make_payment(self):
        pass


class UPI(Payment):

    def __init__(self, user_name, amount):
        super().__init__(user_name, amount)

    def make_payment(self):
        print(f"UPI Payment of ₹{self.amount} successful")


class CreditCard(Payment):

    def __init__(self, user_name, amount):
        super().__init__(user_name, amount)

    def make_payment(self):
        print(f"Credit Card Payment of ₹{self.amount} successful")


u = UPI("Spandan", 5000)
c = CreditCard("Rahul", 10000)

u.make_payment()
c.make_payment()
```

```python
# 7. School Management System

from abc import ABC, abstractmethod

class School(ABC):

    def __init__(self, school_name, city):
        self.school_name = school_name
        self.city = city

    @abstractmethod
    def show_department(self):
        pass


class EngineeringCollege(School):

    def __init__(self, school_name, city):
        super().__init__(school_name, city)

    def show_department(self):
        print("Department : Electronics & Telecommunication")


college = EngineeringCollege("ABC Engineering College", "Hyderabad")

print("College :", college.school_name)
print("City :", college.city)
college.show_department()
```

```python
# 8. Food Delivery App

from abc import ABC, abstractmethod

class FoodDelivery(ABC):

    def __init__(self, customer_name, order_id):
        self.customer_name = customer_name
        self.order_id = order_id

    @abstractmethod
    def place_order(self):
        pass

    @abstractmethod
    def track_order(self):
        pass


class Swiggy(FoodDelivery):

    def __init__(self, customer_name, order_id):
        super().__init__(customer_name, order_id)

    def place_order(self):
        print("Order placed successfully")

    def track_order(self):
        print("Order is on the way")


obj = Swiggy("Spandan", 101)

print("Customer :", obj.customer_name)
print("Order ID :", obj.order_id)

obj.place_order()
obj.track_order()
```

```python
# 9. AI Model Prediction System

from abc import ABC, abstractmethod

class Model(ABC):

    def __init__(self, model_name, version):
        self.model_name = model_name
        self.version = version

    @abstractmethod
    def predict(self):
        pass


class RecommendationModel(Model):

    def __init__(self, model_name, version):
        super().__init__(model_name, version)

    def predict(self):
        print("Recommended products generated")


m = RecommendationModel("Netflix Recommender", "1.0")

print("Model Name :", m.model_name)
print("Version :", m.version)

m.predict()
```

```python
# 10. Hospital Management System

from abc import ABC, abstractmethod

class Hospital(ABC):

    def __init__(self, hospital_name, location):
        self.hospital_name = hospital_name
        self.location = location

    @abstractmethod
    def show_specialization(self):
        pass


class CardiologyDepartment(Hospital):

    def __init__(self, hospital_name, location):
        super().__init__(hospital_name, location)

    def show_specialization(self):
        print("Specialization : Heart Care & Cardiology")


h = CardiologyDepartment("Apollo Hospital", "Hyderabad")

print("Hospital :", h.hospital_name)
print("Location :", h.location)

h.show_specialization()
```
