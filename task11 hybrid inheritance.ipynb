```python
# Company → HR + IT → Employee

class Company:
    def __init__(self, company_name, company_city):
        self.company_name = company_name
        self.company_city = company_city

class HR(Company):
    def __init__(self, company_name, company_city, hr_manager):
        Company.__init__(self, company_name, company_city)
        self.hr_manager = hr_manager

class IT(Company):
    def __init__(self, company_name, company_city, programming_language):
        Company.__init__(self, company_name, company_city)
        self.programming_language = programming_language

class Employee(HR, IT):
    def __init__(self, company_name, company_city, hr_manager,
                 programming_language, employee_name):
        HR.__init__(self, company_name, company_city, hr_manager)
        IT.__init__(self, company_name, company_city, programming_language)
        self.employee_name = employee_name

    def display(self):
        print(self.company_name)
        print(self.company_city)
        print(self.hr_manager)
        print(self.programming_language)
        print(self.employee_name)

e = Employee("Infosys", "Hyderabad", "Ramesh", "Python", "Spandan")
e.display()
```

```python
# Hospital → Doctor + Nurse → Patient

class Hospital:
    def __init__(self, hospital_name, location):
        self.hospital_name = hospital_name
        self.location = location

class Doctor(Hospital):
    def __init__(self, hospital_name, location, doctor_name):
        Hospital.__init__(self, hospital_name, location)
        self.doctor_name = doctor_name

class Nurse(Hospital):
    def __init__(self, hospital_name, location, nurse_name):
        Hospital.__init__(self, hospital_name, location)
        self.nurse_name = nurse_name

class Patient(Doctor, Nurse):
    def __init__(self, hospital_name, location, doctor_name,
                 nurse_name, patient_name):
        Doctor.__init__(self, hospital_name, location, doctor_name)
        Nurse.__init__(self, hospital_name, location, nurse_name)
        self.patient_name = patient_name

    def display(self):
        print(self.hospital_name)
        print(self.location)
        print(self.doctor_name)
        print(self.nurse_name)
        print(self.patient_name)

p = Patient("Apollo", "Hyderabad", "Dr. Rao", "Priya", "Rahul")
p.display()
```

```python
# University → College + School → Student

class University:
    def __init__(self, university_name, city):
        self.university_name = university_name
        self.city = city

class College(University):
    def __init__(self, university_name, city, college_name):
        University.__init__(self, university_name, city)
        self.college_name = college_name

class School(University):
    def __init__(self, university_name, city, school_name):
        University.__init__(self, university_name, city)
        self.school_name = school_name

class Student(College, School):
    def __init__(self, university_name, city,
                 college_name, school_name, student_name):
        College.__init__(self, university_name, city, college_name)
        School.__init__(self, university_name, city, school_name)
        self.student_name = student_name

    def display(self):
        print(self.university_name)
        print(self.city)
        print(self.college_name)
        print(self.school_name)
        print(self.student_name)

s = Student("OU", "Hyderabad", "CBIT", "DPS", "Spandan")
s.display()
```

```python
# Bank → Savings + Loan → Customer

class Bank:
    def __init__(self, bank_name, branch):
        self.bank_name = bank_name
        self.branch = branch

class Savings(Bank):
    def __init__(self, bank_name, branch, savings_account_no):
        Bank.__init__(self, bank_name, branch)
        self.savings_account_no = savings_account_no

class Loan(Bank):
    def __init__(self, bank_name, branch, loan_amount):
        Bank.__init__(self, bank_name, branch)
        self.loan_amount = loan_amount

class Customer(Savings, Loan):
    def __init__(self, bank_name, branch,
                 savings_account_no, loan_amount, customer_name):
        Savings.__init__(self, bank_name, branch, savings_account_no)
        Loan.__init__(self, bank_name, branch, loan_amount)
        self.customer_name = customer_name

    def display(self):
        print(self.bank_name)
        print(self.branch)
        print(self.savings_account_no)
        print(self.loan_amount)
        print(self.customer_name)

c = Customer("SBI", "Ameerpet", 123456, 500000, "Spandan")
c.display()
```

```python
# Vehicle → Car + Bike → Showroom

class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

class Car(Vehicle):
    def __init__(self, brand, model, car_type):
        Vehicle.__init__(self, brand, model)
        self.car_type = car_type

class Bike(Vehicle):
    def __init__(self, brand, model, bike_type):
        Vehicle.__init__(self, brand, model)
        self.bike_type = bike_type

class Showroom(Car, Bike):
    def __init__(self, brand, model, car_type, bike_type, owner_name):
        Car.__init__(self, brand, model, car_type)
        Bike.__init__(self, brand, model, bike_type)
        self.owner_name = owner_name

    def display(self):
        print(self.brand)
        print(self.model)
        print(self.car_type)
        print(self.bike_type)
        print(self.owner_name)

s = Showroom("Honda", "2025", "SUV", "Sports", "Rajesh")
s.display()
```

```python
# School → Teacher + Library → Student

class School:
    def __init__(self, school_name, city):
        self.school_name = school_name
        self.city = city

class Teacher(School):
    def __init__(self, school_name, city, subject):
        School.__init__(self, school_name, city)
        self.subject = subject

class Library(School):
    def __init__(self, school_name, city, book_count):
        School.__init__(self, school_name, city)
        self.book_count = book_count

class Student(Teacher, Library):
    def __init__(self, school_name, city,
                 subject, book_count, student_name):
        Teacher.__init__(self, school_name, city, subject)
        Library.__init__(self, school_name, city, book_count)
        self.student_name = student_name

    def display(self):
        print(self.school_name)
        print(self.city)
        print(self.subject)
        print(self.book_count)
        print(self.student_name)

s = Student("DAV", "Hyderabad", "Maths", 5000, "Spandan")
s.display()
```

```python
# ECommerce → Product + Seller → Customer

class ECommerce:
    def __init__(self, website_name, country):
        self.website_name = website_name
        self.country = country

class Product(ECommerce):
    def __init__(self, website_name, country, product_name):
        ECommerce.__init__(self, website_name, country)
        self.product_name = product_name

class Seller(ECommerce):
    def __init__(self, website_name, country, seller_name):
        ECommerce.__init__(self, website_name, country)
        self.seller_name = seller_name

class Customer(Product, Seller):
    def __init__(self, website_name, country,
                 product_name, seller_name, customer_name):
        Product.__init__(self, website_name, country, product_name)
        Seller.__init__(self, website_name, country, seller_name)
        self.customer_name = customer_name

    def display(self):
        print(self.website_name)
        print(self.country)
        print(self.product_name)
        print(self.seller_name)
        print(self.customer_name)

c = Customer("Amazon", "India", "Laptop", "TechStore", "Spandan")
c.display()
```

```python
# CricketAcademy → Coach + Ground → Player

class CricketAcademy:
    def __init__(self, academy_name, location):
        self.academy_name = academy_name
        self.location = location

class Coach(CricketAcademy):
    def __init__(self, academy_name, location, coach_name):
        CricketAcademy.__init__(self, academy_name, location)
        self.coach_name = coach_name

class Ground(CricketAcademy):
    def __init__(self, academy_name, location, ground_name):
        CricketAcademy.__init__(self, academy_name, location)
        self.ground_name = ground_name

class Player(Coach, Ground):
    def __init__(self, academy_name, location,
                 coach_name, ground_name, player_name):
        Coach.__init__(self, academy_name, location, coach_name)
        Ground.__init__(self, academy_name, location, ground_name)
        self.player_name = player_name

    def display(self):
        print(self.academy_name)
        print(self.location)
        print(self.coach_name)
        print(self.ground_name)
        print(self.player_name)

p = Player("National Academy", "Hyderabad",
           "Rahul Dravid", "ACA Ground", "Virat")
p.display()
```

```python
# Abstract Bank → ATM + MobileBanking → User

from abc import ABC, abstractmethod

class Bank(ABC):
    def __init__(self, bank_name, branch):
        self.bank_name = bank_name
        self.branch = branch

    @abstractmethod
    def show_bank_details(self):
        pass


class ATM(Bank):
    def __init__(self, bank_name, branch, atm_id):
        Bank.__init__(self, bank_name, branch)
        self.atm_id = atm_id


class MobileBanking(Bank):
    def __init__(self, bank_name, branch, mobile_number):
        Bank.__init__(self, bank_name, branch)
        self.mobile_number = mobile_number


class User(ATM, MobileBanking):
    def __init__(self, bank_name, branch, atm_id,
                 mobile_number, user_name):

        ATM.__init__(self, bank_name, branch, atm_id)
        self.mobile_number = mobile_number
        self.user_name = user_name

    def show_bank_details(self):
        print("Bank Name     :", self.bank_name)
        print("Branch        :", self.branch)
        print("ATM ID        :", self.atm_id)
        print("Mobile Number :", self.mobile_number)
        print("User Name     :", self.user_name)


# Create Object
u1 = User(
    "SBI",
    "Ameerpet",
    "ATM101",
    "9876543210",
    "Spandan"
)

u1.show_bank_details()
```
