```python
nums = [1, 2, 3, 4, 5, 6, 7]

result = []

for num in nums:
    if num % 2 != 0:
        result.append(num ** 2)

print(result)
```

```python
nums = [1, 2, 2, 3, 4, 4, 5, 1]

duplicates = []

for num in nums:
    if nums.count(num) > 1 and num not in duplicates:
        duplicates.append(num)

print(duplicates)
```

```python
nums = [11, 22, 33, 44, 55, 66]

even = 0
odd = 0

for num in nums:
    if num % 2 == 0:
        even += 1
    else:
        odd += 1

print("Even:", even)
print("Odd :", odd)
```

```python
nums = [12, 45, 67, 89, 34, 89, 76]

largest = second = float("-inf")

for num in nums:
    if num > largest:
        second = largest
        largest = num
    elif num > second and num != largest:
        second = num

print("Second Largest:", second)
```
