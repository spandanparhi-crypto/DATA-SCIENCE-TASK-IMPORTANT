```python
# 1. Student Management System (Single Inheritance)

class Student:

    def __init__(self, student_name, roll_no):
        self.student_name = student_name
        self.roll_no = roll_no

    def display_student(self):
        print("Student Name :", self.student_name)
        print("Roll No      :", self.roll_no)


class CollegeStudent(Student):

    def __init__(self, student_name, roll_no, course_name):
        super().__init__(student_name, roll_no)
        self.course_name = course_name

    def display_course(self):
        self.display_student()
        print("Course Name  :", self.course_name)


s = CollegeStudent("Spandan", 101, "Data Science")
s.display_course()
```

```python
# 2. Vehicle Information (Without super)

class Vehicle:

    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def show_vehicle(self):
        print("Brand :", self.brand)
        print("Model :", self.model)


class Bike(Vehicle):

    def __init__(self, brand, model, mileage):
        Vehicle.__init__(self, brand, model)
        self.mileage = mileage

    def show_bike(self):
        self.show_vehicle()
        print("Mileage :", self.mileage)


b = Bike("Honda", "Shine", 65)
b.show_bike()
```

```python
# 3. Employee Management

class Employee:

    def __init__(self, employee_name, salary):
        self.employee_name = employee_name
        self.salary = salary

    def display_employee(self):
        print("Employee Name :", self.employee_name)
        print("Salary        :", self.salary)


class Developer(Employee):

    def __init__(self, employee_name, salary, programming_language):
        super().__init__(employee_name, salary)
        self.programming_language = programming_language

    def display_developer(self):
        self.display_employee()
        print("Language      :", self.programming_language)


d = Developer("Spandan", 50000, "Python")
d.display_developer()
```

```python
# 4. Bank System with Abstraction


from abc import ABC, abstractmethod

class Bank(ABC):

    def __init__(self, account_holder, account_number):
        self.account_holder = account_holder
        self.account_number = account_number

    @abstractmethod
    def interest_rate(self):
        pass


class SBI(Bank):

    def __init__(self, account_holder, account_number):
        super().__init__(account_holder, account_number)

    def interest_rate(self):
        print("Interest Rate : 6.5%")


s = SBI("Spandan", 12345)

print("Holder :", s.account_holder)
print("Account:", s.account_number)
s.interest_rate()
```

```python
# 5. Animal Information

class Animal:

    def __init__(self, animal_name):
        self.animal_name = animal_name

    def eat(self):
        print(self.animal_name, "is eating")


class Dog(Animal):

    def bark(self):
        print("Dog is barking")


d = Dog("Tommy")

d.eat()
d.bark()
```

```python
# 6. School Management System (Multilevel)

class Person:

    def __init__(self, name):
        self.name = name


class Student(Person):

    def __init__(self, name, roll_no):
        super().__init__(name)
        self.roll_no = roll_no


class CollegeStudent(Student):

    def __init__(self, name, roll_no, course_name):
        super().__init__(name, roll_no)
        self.course_name = course_name

    def display(self):
        print("Name       :", self.name)
        print("Roll No    :", self.roll_no)
        print("Course     :", self.course_name)


c = CollegeStudent("Spandan", 101, "Data Science")
c.display()
```

```python
# 7. Company Hierarchy


class Company:

    def __init__(self, company_name):
        self.company_name = company_name


class Department(Company):

    def __init__(self, company_name, department_name):
        super().__init__(company_name)
        self.department_name = department_name


class Employee(Department):

    def __init__(self, company_name, department_name, employee_name):
        super().__init__(company_name, department_name)
        self.employee_name = employee_name

    def display(self):
        print("Company    :", self.company_name)
        print("Department :", self.department_name)
        print("Employee   :", self.employee_name)


e = Employee("TCS", "AI", "Spandan")
e.display()
```

```python
# 8. Banking System (Abstraction + Multilevel)

from abc import ABC, abstractmethod

class Bank(ABC):

    def __init__(self, bank_name):
        self.bank_name = bank_name

    @abstractmethod
    def show_interest(self):
        pass


class SBI(Bank):

    def __init__(self, bank_name, branch_name):
        super().__init__(bank_name)
        self.branch_name = branch_name

    def show_interest(self):
        print("Interest Rate : 6.5%")


class Customer(SBI):

    def __init__(self, bank_name, branch_name, customer_name):
        super().__init__(bank_name, branch_name)
        self.customer_name = customer_name

    def display(self):
        print("Bank Name     :", self.bank_name)
        print("Branch Name   :", self.branch_name)
        print("Customer Name :", self.customer_name)
        self.show_interest()


c = Customer("SBI", "Ameerpet", "Spandan")
c.display()
```

```python
# 9. E-Commerce System (Without super)

class Product:

    def __init__(self, product_name):
        self.product_name = product_name


class Electronics(Product):

    def __init__(self, product_name, warranty):
        Product.__init__(self, product_name)
        self.warranty = warranty


class Mobile(Electronics):

    def __init__(self, product_name, warranty, price):
        Electronics.__init__(self, product_name, warranty)
        self.price = price

    def display(self):
        print("Product  :", self.product_name)
        print("Warranty :", self.warranty)
        print("Price    :", self.price)


m = Mobile("Samsung M35", "2 Years", 25000)
m.display()
```

```python
# 10. Hospital Management System (Hard)

from abc import ABC, abstractmethod

class Hospital(ABC):

    def __init__(self, hospital_name):
        self.hospital_name = hospital_name

    @abstractmethod
    def show_specialization(self):
        pass


class Doctor(Hospital):

    def __init__(self, hospital_name, doctor_name):
        super().__init__(hospital_name)
        self.doctor_name = doctor_name

    def show_specialization(self):
        print("General Doctor")


class Cardiologist(Doctor):

    def __init__(self, hospital_name, doctor_name, experience):
        super().__init__(hospital_name, doctor_name)
        self.experience = experience

    def show_specialization(self):
        print("Specialization : Cardiology")

    def show_all_details(self):
        print("Hospital Name :", self.hospital_name)
        print("Doctor Name   :", self.doctor_name)
        print("Experience    :", self.experience, "Years")
        self.show_specialization()


c = Cardiologist("Apollo Hospital", "Dr. spandan", 12)
c.show_all_details()


```
