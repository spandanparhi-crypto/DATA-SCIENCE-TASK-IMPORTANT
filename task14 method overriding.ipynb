```python
# Q1 — Animal Method Overriding

class Animal:
    def speak(self):
        print("Animal makes a sound")


class Dog(Animal):
    def speak(self):
        print("Dog says: Woof Woof")


class Cat(Animal):
    def speak(self):
        print("Cat says: Meow")


class Cow(Animal):
    def speak(self):
        print("Cow says: Moo")


class Sheep(Animal):
    def speak(self):
        print("Sheep says: Baa")


animals = [Dog(), Cat(), Cow(), Sheep()]

for animal in animals:
    animal.speak()
```

```python
# Q2 — Shape Method Overriding

class Shape:
    def draw(self):
        print("Drawing a shape")


class Circle(Shape):
    def draw(self):
        print("⭕ Circle Drawn")


class Rectangle(Shape):
    def draw(self):
        print("▭ Rectangle Drawn")


class Triangle(Shape):
    def draw(self):
        print("🔺 Triangle Drawn")


class Square(Shape):
    def draw(self):
        print("⬜ Square Drawn")


shapes = [Circle(), Rectangle(), Triangle(), Square()]

for shape in shapes:
    shape.draw()
```

```python
# Q3 — Phone Method Overriding

class Phone:
    def call(self):
        print("Calling...")

    def message(self):
        print("Sending Message...")


class Samsung(Phone):
    def call(self):
        print("Samsung HD Calling")

    def message(self):
        print("Samsung Message Sent")


class iPhone(Phone):
    def call(self):
        print("iPhone FaceTime Audio Call")

    def message(self):
        print("iMessage Sent")


class OnePlus(Phone):
    def call(self):
        print("OnePlus Clear Calling")

    def message(self):
        print("OnePlus Message Sent")


class Nokia(Phone):
    def call(self):
        print("Nokia Classic Calling")

    def message(self):
        print("Nokia SMS Sent")


phones = [Samsung(), iPhone(), OnePlus(), Nokia()]

for phone in phones:
    phone.call()
    phone.message()
    print("-" * 25)
```

```python
# Q4 — Vehicle Method Overriding

class Vehicle:
    def fuel_type(self):
        print("Uses fuel")

    def move(self):
        print("Vehicle Moving")

    def show(self):
        self.fuel_type()
        self.move()
        print("-" * 30)


class Car(Vehicle):
    def fuel_type(self):
        print("Fuel: Petrol")

    def move(self):
        print("Car moves on road")


class Bike(Vehicle):
    def fuel_type(self):
        print("Fuel: Petrol")

    def move(self):
        print("Bike moves on road")


class Boat(Vehicle):
    def fuel_type(self):
        print("Fuel: Diesel")

    def move(self):
        print("Boat moves on water")


class Airplane(Vehicle):
    def fuel_type(self):
        print("Fuel: Aviation Fuel")

    def move(self):
        print("Airplane flies in sky")


class Rocket(Vehicle):
    def fuel_type(self):
        print("Fuel: Liquid Hydrogen")

    def move(self):
        print("Rocket moves into space")


vehicles = [Car(), Bike(), Boat(), Airplane(), Rocket()]

for v in vehicles:
    v.show()
```

```python
# Q5 — Employee Method Overriding

class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def work(self):
        print("Employee is working")

    def calculate_bonus(self):
        return self.salary * 0.05


class Developer(Employee):
    def work(self):
        print("Writing Code")

    def calculate_bonus(self):
        return self.salary * 0.20


class Designer(Employee):
    def work(self):
        print("Creating UI")

    def calculate_bonus(self):
        return self.salary * 0.15


class Manager(Employee):
    def work(self):
        print("Managing Team")

    def calculate_bonus(self):
        return self.salary * 0.25


class Intern(Employee):
    def work(self):
        print("Learning")

    def calculate_bonus(self):
        return self.salary * 0.05


employees = [
    Developer("Rahul", 50000),
    Designer("Priya", 40000),
    Manager("Amit", 80000),
    Intern("Riya", 15000)
]

highest = employees[0]

for emp in employees:
    print(emp.name)
    emp.work()
    print("Bonus:", emp.calculate_bonus())
    print()

    if emp.calculate_bonus() > highest.calculate_bonus():
        highest = emp

print("Highest Bonus:", highest.name, highest.calculate_bonus())
```

```python
# Q6 — Food Method Overriding

class Food:
    def prepare(self):
        print("Preparing food")

    def serve(self):
        print("Serving food")

    def order(self):
        self.prepare()
        self.serve()
        print("-" * 25)


class Pizza(Food):
    def prepare(self):
        print("Preparing Pizza")

    def serve(self):
        print("Serving Pizza")


class Burger(Food):
    def prepare(self):
        print("Preparing Burger")

    def serve(self):
        print("Serving Burger")


class Sushi(Food):
    def prepare(self):
        print("Preparing Sushi")

    def serve(self):
        print("Serving Sushi")


class Pasta(Food):
    def prepare(self):
        print("Preparing Pasta")

    def serve(self):
        print("Serving Pasta")


foods = [
    Pizza(), Pizza(),
    Burger(), Burger(),
    Sushi(), Sushi(),
    Pasta(), Pasta()
]

for food in foods:
    food.order()
```

```python
# Q7 — Bank Method Overriding

class Bank:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def account_type(self):
        return "Basic Account"

    def interest_rate(self):
        return 0

    def show(self):
        print("Owner :", self.owner)
        print("Balance :", self.balance)
        print("Type :", self.account_type())
        print("Interest :", self.interest_rate(), "%")
        print("-" * 30)


class SavingsAccount(Bank):
    def account_type(self):
        return "Savings Account"

    def interest_rate(self):
        return 4


class CurrentAccount(Bank):
    def account_type(self):
        return "Current Account"

    def interest_rate(self):
        return 1


class FixedDeposit(Bank):
    def account_type(self):
        return "Fixed Deposit"

    def interest_rate(self):
        return 7


class LoanAccount(Bank):
    def account_type(self):
        return "Loan Account"

    def interest_rate(self):
        return 10


accounts = [
    SavingsAccount("Rahul", 50000),
    CurrentAccount("Priya", 70000),
    FixedDeposit("Amit", 100000),
    LoanAccount("Riya", 200000)
]

for acc in accounts:
    acc.show()
```

```python
# Q8 — Person Method Overriding

class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    def get_name(self):
        return self.__name

    def get_age(self):
        return self.__age

    def introduce(self):
        print("I am", self.__name)

    def show_role(self):
        print("Person")


class Student(Person):
    def introduce(self):
        print("I am a Student")

    def show_role(self):
        print("Studying")


class Teacher(Person):
    def introduce(self):
        print("I am a Teacher")

    def show_role(self):
        print("Teaching")


class Doctor(Person):
    def introduce(self):
        print("I am a Doctor")

    def show_role(self):
        print("Treating Patients")


class Engineer(Person):
    def introduce(self):
        print("I am an Engineer")

    def show_role(self):
        print("Building Systems")


people = [
    Student("Rahul", 20),
    Teacher("Priya", 35),
    Doctor("Amit", 40),
    Engineer("Riya", 28)
]

for p in people:
    p.introduce()
    p.show_role()
    print()
```

```python
# Q9 — Game Method Overriding

class Game:
    total_games = 0

    def __init__(self):
        Game.total_games += 1

    def start(self):
        print("Game Starting...")

    def play(self):
        print("Playing Game")

    def score(self):
        return 0

    def end(self):
        print("Game Over")


class Cricket(Game):
    def start(self):
        print("Cricket Match Started")

    def play(self):
        print("Batting and Bowling")

    def score(self):
        return 250

    def end(self):
        print("Cricket Match Ended")


class Football(Game):
    def start(self):
        print("Football Match Started")

    def play(self):
        print("Kicking Ball")

    def score(self):
        return 3

    def end(self):
        print("Football Match Ended")


class Basketball(Game):
    def start(self):
        print("Basketball Match Started")

    def play(self):
        print("Shooting Hoops")

    def score(self):
        return 120

    def end(self):
        print("Basketball Match Ended")


class Tennis(Game):
    def start(self):
        print("Tennis Match Started")

    def play(self):
        print("Serving Ball")

    def score(self):
        return 6

    def end(self):
        print("Tennis Match Ended")


games = [Cricket(), Football(), Basketball(), Tennis()]

for g in games:
    g.start()
    g.play()
    print("Score:", g.score())
    g.end()
    print()

print("Total Games:", Game.total_games)
```
