```python
#  student Name Update


class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

student = Student("Rahul", 20)

student.name = "Spandan"

print("Updated Name:", student.name)
```

```python
# Car Speed Check

class Car:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

car = Car("Toyota", 90)

if car.speed > 80:
    print("Over Speed")
else:
    print("Safe Driving")
```

```python
# Employee Salary Update

class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

emp = Employee("Rahul", 25000)

emp.salary += 5000

print("Updated Salary:", emp.salary)
```

```python
# Product Stock Check

class Product:
    def __init__(self, product_name, stock):
        self.product_name = product_name
        self.stock = stock

product = Product("Laptop", 0)

if product.stock == 0:
    print("Out of Stock")
else:
    print("Available")
```

```python
# Mobile Battery Status


class Mobile:
    def __init__(self, model, battery):
        self.model = model
        self.battery = battery

mobile = Mobile("Samsung", 15)

if mobile.battery < 20:
    print("Low Battery")
elif mobile.battery <= 50:
    print("Medium Level")
else:
    print("Battery OK")
```

```python
# Bank Account Deposit

class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

account = BankAccount("Rahul", 18000)

account.balance += 5000

print("Updated Balance:", account.balance)

if account.balance > 20000:
    print("Premium Customer")
else:
    print("Regular Customer")
```

```python
#  Student Result System

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

student = Student("Rahul", 82)

if student.marks >= 90:
    print("Grade A")
elif student.marks >= 75:
    print("Grade B")
elif student.marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

```python
# Laptop Price Discount

class Laptop:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

laptop = Laptop("HP", 60000)

if laptop.price > 50000:
    discount = laptop.price * 0.10
    laptop.price -= discount

print("Updated Price:", laptop.price)
```

```python
# Movie Ticket Booking


class MovieTicket:
    def __init__(self, name, age):
        self.name = name
        self.age = age

person = MovieTicket("Rahul", 17)

if person.age >= 18:
    print("Allowed")
else:
    print("Not Allowed")

person.age += 1

print("Updated Age:", person.age)
```

```python
# Online Shopping Cart

class ShoppingCart:
    def __init__(self, customer_name, total_amount):
        self.customer_name = customer_name
        self.total_amount = total_amount

cart = ShoppingCart("Rahul", 1500)

if cart.total_amount > 1000:
    discount = cart.total_amount * 0.20
else:
    discount = cart.total_amount * 0.05

final_amount = cart.total_amount - discount

print("Original Amount:", cart.total_amount)
print("Discount:", discount)
print("Final Amount:", final_amount)
```
