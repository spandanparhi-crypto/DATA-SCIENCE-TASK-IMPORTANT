```python
# 1. Handle ValueError

try:
    num = int(input("Enter a number: "))
    print("Number =", num)
except ValueError:
    print("Please enter a valid number!")
```

```python
# 2. Handle ZeroDivisionError
try:
    a = int(input("Enter numerator: "))
    b = int(input("Enter denominator: "))

    result = a / b
    print("Result =", result)

except ZeroDivisionError:
    print("Cannot divide by zero!")
```

```python
# 3. Multiple Exceptions

try:
    num = int(input("Enter a number: "))
    print("Result =", 100 / num)

except ValueError:
    print("Please enter a valid number!")

except ZeroDivisionError:
    print("Cannot divide by zero!")
```

```python
# 4. Print Actual Error Message

try:
    num = int("hello")

except ValueError as e:
    print("Error:", e)
```

```python
# 5. File Not Found
try:
    with open("student.txt", "r") as file:
        print(file.read())

except FileNotFoundError:
    print("File not found!")
```

```python
# 6. Use else Block

try:
    num = int(input("Enter a number: "))

except ValueError:
    print("Invalid Input")

else:
    print("Square =", num ** 2)
```

```python
# 7. Use finally Block

try:
    with open("student.txt", "r") as file:
        print(file.read())

except FileNotFoundError:
    print("File not found!")

finally:
    print("Program Finished")
```

```python
# 8. Catch Multiple Errors in One Line

try:
    num = int(input("Enter a number: "))
    print(100 / num)

except (ValueError, ZeroDivisionError):
    print("Error occurred!")
```

```python
# 9. NameError Handling

try:
    print(name)

except NameError:
    print("Variable not defined!")
```

```python
# 10. TypeError Handling

try:
    result = 10 + "Python"
    print(result)

except TypeError:
    print("Cannot add integer and string!")
```

```python
# 11. IndexError Handling

try:
    lst = [10, 20, 30, 40, 50]
    print(lst[10])

except IndexError:
    print("Index out of range!")
```

```python
# 12. KeyError Handling

try:
    student = {"name": "Rahul"}

    print(student["age"])

except KeyError:
    print("Key not found!")
```

```python
# 13. UnboundLocalError

def demo():
    try:
        print(x)
        x = 10

    except UnboundLocalError:
        print("Local variable used before assignment!")

demo()
```

```python
# 14. Generic Exception

try:
    num = int(input("Enter number: "))
    print(100 / num)

except Exception as e:
    print("Error:", e)
```

```python
# 15. Complete Exception Handling Program

try:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))

    result = a / b

except ValueError:
    print("Please enter valid numbers!")

except ZeroDivisionError:
    print("Cannot divide by zero!")

else:
    print("Result =", result)

finally:
    print("Program Finished")
```

```python
# Bonus Challenge: Bank Withdrawal System

class InsufficientBalanceError(Exception):
    pass

balance = 5000

try:
    amount = int(input("Enter withdrawal amount: "))

    if amount > balance:
        raise InsufficientBalanceError(
            "Insufficient Balance!"
        )

except ValueError:
    print("Please enter a valid amount!")

except InsufficientBalanceError as e:
    print(e)

else:
    balance -= amount
    print("Withdrawal Successful")
    print("Remaining Balance =", balance)

finally:
    print("Transaction Completed")
```
