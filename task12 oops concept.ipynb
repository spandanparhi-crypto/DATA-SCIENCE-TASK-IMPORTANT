```python
# 3. Bank Account System

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        self.__balance -= amount

    def show_balance(self):
        print("Balance:", self.__balance)


b1 = BankAccount(10000)
b1.deposit(2000)
b1.withdraw(1000)
b1.show_balance()
```

```python
# 4. Student Marks Validation

class Student:
    def __init__(self):
        self.__marks = 0

    def set_marks(self, marks):
        if 0 <= marks <= 100:
            self.__marks = marks
            print("Marks Updated")
        else:
            print("Invalid Marks")

    def display_marks(self):
        print("Marks:", self.__marks)


s1 = Student()
s1.set_marks(85)
s1.display_marks()
```

```python
# 6. ATM System


class ATM:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount
        print(amount, "deposited successfully")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(amount, "withdrawn successfully")
        else:
            print("Insufficient Balance")

    def show_balance(self):
        print("Current Balance:", self.__balance)


atm = ATM(5000)
atm.deposit(2000)
atm.withdraw(3000)
atm.show_balance()
```

```python
# 7. User Password System

class User:
    def __init__(self, password):
        self.__password = password

    def change_password(self, old_password, new_password):
        if self.__password == old_password:
            self.__password = new_password
            print("Password Changed Successfully")
        else:
            print("Incorrect Old Password")

    def validate_password(self, password):
        if self.__password == password:
            print("Password Valid")
        else:
            print("Invalid Password")


u1 = User("admin123")
u1.validate_password("admin123")
u1.change_password("admin123", "python123")
u1.validate_password("python123")

```

```python
# 8. Employee → Developer (Single Inheritance)

class Employee:
    def __init__(self, employee_name, employee_id):
        self.employee_name = employee_name
        self.employee_id = employee_id


class Developer(Employee):
    def __init__(self, employee_name, employee_id, programming_language):
        super().__init__(employee_name, employee_id)
        self.programming_language = programming_language

    def display(self):
        print("Employee Name:", self.employee_name)
        print("Employee ID:", self.employee_id)
        print("Programming Language:", self.programming_language)


d1 = Developer("Spandan", 101, "Python")
d1.display()
```

```python
# 9. Vehicle → Car (Single Inheritance)

class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model


class Car(Vehicle):
    def __init__(self, brand, model, fuel_type):
        super().__init__(brand, model)
        self.fuel_type = fuel_type

    def display(self):
        print("Brand:", self.brand)
        print("Model:", self.model)
        print("Fuel Type:", self.fuel_type)


c1 = Car("Toyota", "Innova", "Diesel")
c1.display()
```

```python
# 11. Person → Employee → Manager (Multilevel Inheritance)

class Person:
    def __init__(self, name):
        self.name = name


class Employee(Person):
    def __init__(self, name, employee_id):
        super().__init__(name)
        self.employee_id = employee_id


class Manager(Employee):
    def __init__(self, name, employee_id, department):
        super().__init__(name, employee_id)
        self.department = department

    def display(self):
        print("Name:", self.name)
        print("Employee ID:", self.employee_id)
        print("Department:", self.department)


m1 = Manager("Spandan", 101, "IT")
m1.display()
```

```python
# 12. Company → Department → Employee (Multilevel Inheritance)


class Company:
    def __init__(self, company_name):
        self.company_name = company_name


class Department(Company):
    def __init__(self, company_name, department_name):
        super().__init__(company_name)
        self.department_name = department_name


class Employee(Department):
    def __init__(self, company_name, department_name, employee_name):
        super().__init__(company_name, department_name)
        self.employee_name = employee_name

    def display(self):
        print("Company Name:", self.company_name)
        print("Department Name:", self.department_name)
        print("Employee Name:", self.employee_name)


e1 = Employee("Infosys", "Data Science", "Spandan")
e1.display()
```

```python
# 13. University → College → Student (Multilevel Inheritance)

class University:
    def __init__(self, university_name):
        self.university_name = university_name


class College(University):
    def __init__(self, university_name, college_name):
        super().__init__(university_name)
        self.college_name = college_name


class Student(College):
    def __init__(self, university_name, college_name, student_name):
        super().__init__(university_name, college_name)
        self.student_name = student_name

    def display(self):
        print("University Name:", self.university_name)
        print("College Name:", self.college_name)
        print("Student Name:", self.student_name)


s1 = Student("BPUT", "BCET", "Spandan")
s1.display()
```

```python
# 14. Father + Mother → Child (Multiple Inheritance)

class Father:
    def __init__(self, father_name, father_job):
        self.father_name = father_name
        self.father_job = father_job


class Mother:
    def __init__(self, mother_name, mother_job):
        self.mother_name = mother_name
        self.mother_job = mother_job


class Child(Father, Mother):
    def __init__(self, father_name, father_job,
                 mother_name, mother_job,
                 child_name):
        Father.__init__(self, father_name, father_job)
        Mother.__init__(self, mother_name, mother_job)
        self.child_name = child_name

    def display(self):
        print("Father Name:", self.father_name)
        print("Father Job:", self.father_job)
        print("Mother Name:", self.mother_name)
        print("Mother Job:", self.mother_job)
        print("Child Name:", self.child_name)


c1 = Child("Ramesh", "Engineer",
           "Sita", "Teacher",
           "Spandan")
c1.display()
```

```python
# 15. Teacher + Engineer → Trainer (Multiple Inheritance)

class Teacher:
    def __init__(self, subject):
        self.subject = subject


class Engineer:
    def __init__(self, programming_language):
        self.programming_language = programming_language


class Trainer(Teacher, Engineer):
    def __init__(self, subject, programming_language, trainer_name):
        Teacher.__init__(self, subject)
        Engineer.__init__(self, programming_language)
        self.trainer_name = trainer_name

    def display(self):
        print("Trainer Name:", self.trainer_name)
        print("Subject:", self.subject)
        print("Programming Language:", self.programming_language)


t1 = Trainer("Python", "Python", "Spandan")
t1.display()
```

```python
# 16. BankAccount + Loan → Customer (Multiple Inheritance)

class BankAccount:
    def __init__(self, account_number):
        self.account_number = account_number


class Loan:
    def __init__(self, loan_amount):
        self.loan_amount = loan_amount


class Customer(BankAccount, Loan):
    def __init__(self, account_number, loan_amount, customer_name):
        BankAccount.__init__(self, account_number)
        Loan.__init__(self, loan_amount)
        self.customer_name = customer_name

    def display(self):
        print("Customer Name:", self.customer_name)
        print("Account Number:", self.account_number)
        print("Loan Amount:", self.loan_amount)


c1 = Customer("ACC1001", 500000, "Spandan")
c1.display()
```

```python
# 17. University → College + School → Student (Hybrid Inheritance)

class University:
    def __init__(self, university_name, city):
        self.university_name = university_name
        self.city = city


class College(University):
    def __init__(self, university_name, city, college_name):
        # Use explicit parent class call instead of super() to avoid MRO conflicts
        University.__init__(self, university_name, city)
        self.college_name = college_name


class School(University):
    def __init__(self, university_name, city, school_name):
        # Use explicit parent class call instead of super() to avoid MRO conflicts
        University.__init__(self, university_name, city)
        self.school_name = school_name


class Student(College, School):
    def __init__(self, university_name, city,
                 college_name, school_name, student_name):

        # Explicitly call both parent class constructors
        College.__init__(self, university_name, city, college_name)
        School.__init__(self, university_name, city, school_name)

        self.student_name = student_name

    def display(self):
        print("University Name :", self.university_name)
        print("City            :", self.city)
        print("College Name    :", self.college_name)
        print("School Name     :", self.school_name)
        print("Student Name    :", self.student_name)


# Create Object
s1 = Student(
    "BPUT",
    "Rourkela",
    "BCET",
    "DAV School",
    "Spandan"
)

s1.display()
```

```python
# 18. Hospital → Doctor + Nurse → Patient (Hybrid Inheritance)
class Hospital:
    def __init__(self, hospital_name):
        self.hospital_name = hospital_name


class Doctor(Hospital):
    def __init__(self, hospital_name, doctor_name):
        Hospital.__init__(self, hospital_name)
        self.doctor_name = doctor_name


class Nurse(Hospital):
    def __init__(self, hospital_name, nurse_name):
        Hospital.__init__(self, hospital_name)
        self.nurse_name = nurse_name


class Patient(Doctor, Nurse):
    def __init__(self, hospital_name, doctor_name, nurse_name, patient_name):
        Doctor.__init__(self, hospital_name, doctor_name)
        Nurse.__init__(self, hospital_name, nurse_name)
        self.patient_name = patient_name

    def show(self):
        print("Hospital Name :", self.hospital_name)
        print("Doctor Name   :", self.doctor_name)
        print("Nurse Name    :", self.nurse_name)
        print("Patient Name  :", self.patient_name)


# Create Object
p1 = Patient(
    "Apollo Hospital",
    "Dr. Sharma",
    "Nurse Priya",
    "Rahul"
)

p1.show()
```

```python
# 19. Bank System (Abstraction)

from abc import ABC, abstractmethod

class Bank(ABC):
    def __init__(self, bank_name, branch):
        self.bank_name = bank_name
        self.branch = branch

    @abstractmethod
    def show_interest_rate(self):
        pass


class SBI(Bank):
    def __init__(self, bank_name, branch):
        super().__init__(bank_name, branch)

    def show_interest_rate(self):
        print("Bank Name:", self.bank_name)
        print("Branch:", self.branch)
        print("Interest Rate: 6.5%")


s1 = SBI("SBI", "Balasore")
s1.show_interest_rate()
```

```python
# 20. University System (Abstraction + Inheritance)

from abc import ABC, abstractmethod

class University(ABC):
    def __init__(self, university_name, city):
        self.university_name = university_name
        self.city = city

    @abstractmethod
    def show_university_details(self):
        pass


class EngineeringCollege(University):
    def __init__(self, university_name, city, college_name):
        super().__init__(university_name, city)
        self.college_name = college_name

    def show_university_details(self):
        print("University Name:", self.university_name)
        print("City:", self.city)
        print("College Name:", self.college_name)


e1 = EngineeringCollege("BPUT", "Rourkela", "BCET")
e1.show_university_details()
```
