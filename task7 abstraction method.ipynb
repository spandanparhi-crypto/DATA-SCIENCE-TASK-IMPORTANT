```python
# 2. Animal Sound


from abc import ABC, abstractmethod

class Animal(ABC):

    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):

    def sound(self):
        print("Dog says Woof")

class Cat(Animal):

    def sound(self):
        print("Cat says Meow")

d = Dog()
c = Cat()

d.sound()
c.sound()
```

```python
# 3. Shape Area Calculation

from abc import ABC, abstractmethod
import math

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):

    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        print("Area =", self.length * self.width)

class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        print("Area =", round(math.pi * self.radius * self.radius, 2))

r = Rectangle(10, 5)
c = Circle(7)

r.area()
c.area()
```

```python
# 4. Payment System

from abc import ABC, abstractmethod

class Payment(ABC):

    @abstractmethod
    def pay(self):
        pass

class CreditCardPayment(Payment):

    def pay(self):
        print("Payment done using Credit Card")

class UpiPayment(Payment):

    def pay(self):
        print("Payment done using UPI")

c = CreditCardPayment()
u = UpiPayment()

c.pay()
u.pay()
```

```python
# 5. Employee Work

from abc import ABC, abstractmethod

class Employee(ABC):

    @abstractmethod
    def work(self):
        pass

class Developer(Employee):

    def work(self):
        print("Developer writes code")

class DataAnalyst(Employee):

    def work(self):
        print("Data Analyst analyzes data")

d = Developer()
a = DataAnalyst()

d.work()
a.work()
```

```python
#  6. Online Food Delivery App

from abc import ABC, abstractmethod

class FoodDelivery(ABC):

    @abstractmethod
    def order_food(self):
        pass

    @abstractmethod
    def track_order(self):
        pass

class Swiggy(FoodDelivery):

    def order_food(self):
        print("Food ordered through Swiggy")

    def track_order(self):
        print("Tracking order in Swiggy")

class Zomato(FoodDelivery):

    def order_food(self):
        print("Food ordered through Zomato")

    def track_order(self):
        print("Tracking order in Zomato")

s = Swiggy()
z = Zomato()

s.order_food()
s.track_order()

z.order_food()
z.track_order()
```

```python
# 7. Bank Transaction System

from abc import ABC, abstractmethod

class BankTransaction(ABC):

    @abstractmethod
    def deposit(self):
        pass

    @abstractmethod
    def withdraw(self):
        pass

class SBI(BankTransaction):

    def deposit(self):
        print("Money deposited in SBI")

    def withdraw(self):
        print("Money withdrawn from SBI")

class HDFC(BankTransaction):

    def deposit(self):
        print("Money deposited in HDFC")

    def withdraw(self):
        print("Money withdrawn from HDFC")

s = SBI()
h = HDFC()

s.deposit()
s.withdraw()

h.deposit()
h.withdraw()
```

```python
# 8. Notification System

from abc import ABC, abstractmethod

class Notification(ABC):

    @abstractmethod
    def send_notification(self):
        pass

class EmailNotification(Notification):

    def send_notification(self):
        print("Email Notification Sent")

class SMSNotification(Notification):

    def send_notification(self):
        print("SMS Notification Sent")

class WhatsAppNotification(Notification):

    def send_notification(self):
        print("WhatsApp Notification Sent")

e = EmailNotification()
s = SMSNotification()
w = WhatsAppNotification()

e.send_notification()
s.send_notification()
w.send_notification()
```

```python
# 9. E-Commerce Shipping

from abc import ABC, abstractmethod

class Shipping(ABC):

    @abstractmethod
    def calculate_charges(self):
        pass

class StandardShipping(Shipping):

    def calculate_charges(self):
        print("Shipping Charge = ₹50")

class ExpressShipping(Shipping):

    def calculate_charges(self):
        print("Shipping Charge = ₹150")

s = StandardShipping()
e = ExpressShipping()

s.calculate_charges()
e.calculate_charges()
```

```python
# 10. AI Model Prediction System

from abc import ABC, abstractmethod

class Model(ABC):

    @abstractmethod
    def predict(self):
        pass

class RecommendationModel(Model):

    def predict(self):
        print("Recommended products generated")

class FraudDetectionModel(Model):

    def predict(self):
        print("Fraud transaction detected")

r = RecommendationModel()
f = FraudDetectionModel()

r.predict()
f.predict()
```
