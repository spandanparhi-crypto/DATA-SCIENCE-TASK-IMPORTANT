```python
# 1. __eq__ (Equal)

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __eq__(self, other):
        return self.marks == other.marks


s1 = Student("Spandan", 90)
s2 = Student("Rahul", 90)

print(s1 == s2)
```

```python
# 2. __lt__ (Less Than)

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __lt__(self, other):
        return self.price < other.price


p1 = Product("Mobile", 15000)
p2 = Product("Laptop", 50000)

print(p1 < p2)
```

```python
# 3. __gt__ (Greater Than)

class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def __gt__(self, other):
        return self.salary > other.salary


e1 = Employee("A", 50000)
e2 = Employee("B", 40000)

print(e1 > e2)
```

```python
# 4. __iadd__

class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def __iadd__(self, amount):
        self.balance += amount
        return self


acc = BankAccount(1000)

acc += 500

print(acc.balance)
```

```python
# 5. __isub__

class Wallet:
    def __init__(self, money):
        self.money = money

    def __isub__(self, amount):
        self.money -= amount
        return self


w = Wallet(1000)

w -= 300

print(w.money)
```

```python
# 6. __len__

class Library:
    def __init__(self, books):
        self.books = books

    def __len__(self):
        return len(self.books)


lib = Library(["Python", "Java", "C"])

print(len(lib))
```

```python
# 7. __getitem__

class Playlist:
    def __init__(self, songs):
        self.songs = songs

    def __getitem__(self, index):
        return self.songs[index]


p = Playlist(["Song1", "Song2", "Song3"])

print(p[0])
```

```python
# 8. __setitem_

class Marksheet:
    def __init__(self, marks):
        self.marks = marks

    def __setitem__(self, index, value):
        self.marks[index] = value


m = Marksheet([70, 80, 90])

m[1] = 95

print(m.marks)
```

```python
# 9. __contains__

class Classroom:
    def __init__(self, students):
        self.students = students

    def __contains__(self, name):
        return name in self.students


c = Classroom(["Rahul", "Spandan", "Amit"])

print("Rahul" in c)
```

```python
# 10. __iter__

class Team:
    def __init__(self, players):
        self.players = players

    def __iter__(self):
        return iter(self.players)


t = Team(["Virat", "Rohit", "Gill"])

for player in t:
    print(player)
```

```python
# 11. __bool__

class Cart:
    def __init__(self, items):
        self.items = items

    def __bool__(self):
        return len(self.items) > 0


c1 = Cart(["Book"])
c2 = Cart([])

print(bool(c1))
print(bool(c2))
```

```python
# 12. __call__

class Calculator:
    def __init__(self, num):
        self.num = num

    def __call__(self, value):
        return self.num * value


c = Calculator(10)

print(c(5))
```

```python
# Q13. MovieCollection (__len__, __getitem__, __contains__)

class MovieCollection:
    def __init__(self, movies):
        self.movies = movies

    def __len__(self):
        return len(self.movies)

    def __getitem__(self, index):
        return self.movies[index]

    def __contains__(self, movie):
        return movie in self.movies


m = MovieCollection(["Avatar", "KGF", "Pushpa"])

print(len(m))
print(m[0])
print("Avatar" in m)
```

```python
# Q14. CricketTeam (__iter__, __len__, __contains__)

class CricketTeam:
    def __init__(self, players):
        self.players = players

    def __iter__(self):
        return iter(self.players)

    def __len__(self):
        return len(self.players)

    def __contains__(self, player):
        return player in self.players


team = CricketTeam(["Virat", "Rohit", "Gill"])

print(len(team))
print("Virat" in team)

for p in team:
    print(p)
```

```python
# Q15. ScoreBoard (__eq__, __gt__, __lt__)

class ScoreBoard:
    def __init__(self, score):
        self.score = score

    def __eq__(self, other):
        return self.score == other.score

    def __gt__(self, other):
        return self.score > other.score

    def __lt__(self, other):
        return self.score < other.score


s1 = ScoreBoard(100)
s2 = ScoreBoard(80)

print(s1 == s2)
print(s1 > s2)
print(s1 < s2)
```

```python
# Q16. Inventory (__getitem__, __setitem__, __len__, __iter__)

class Inventory:
    def __init__(self, items):
        self.items = items

    def __getitem__(self, index):
        return self.items[index]

    def __setitem__(self, index, value):
        self.items[index] = value

    def __len__(self):
        return len(self.items)

    def __iter__(self):
        return iter(self.items)


inv = Inventory(["Mouse", "Keyboard", "Monitor"])

print(inv[0])

inv[1] = "Printer"

print(len(inv))

for item in inv:
    print(item)
```

```python
# Q17. DigitalWallet (__iadd__, __isub__, __bool__, __gt__)

class DigitalWallet:
    def __init__(self, balance):
        self.balance = balance

    def __iadd__(self, amount):
        self.balance += amount
        return self

    def __isub__(self, amount):
        self.balance -= amount
        return self

    def __bool__(self):
        return self.balance > 0

    def __gt__(self, other):
        return self.balance > other.balance


w1 = DigitalWallet(1000)
w2 = DigitalWallet(500)

w1 += 500
print(w1.balance)

w1 -= 300
print(w1.balance)

print(bool(w1))

print(w1 > w2)
```
