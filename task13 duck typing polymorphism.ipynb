```python
# Class 1
class CEO:

    def des(self):
        print("Designation : CEO")

    def work(self):
        print("CEO manages the entire company")

    def salary(self):
        print("Salary : ₹2,50,000")


# Class 2
class Manager:

    def des(self):
        print("Designation : Manager")

    def work(self):
        print("Manager manages teams and projects")

    def salary(self):
        print("Salary : ₹80,000")


# Class 3
class Employee:

    def des(self):
        print("Designation : Employee")

    def work(self):
        print("Employee completes assigned tasks")

    def salary(self):
        print("Salary : ₹30,000")


# Duck Typing Function
def show_details(person):
    person.des()
    person.work()
    person.salary()
    print("-" * 30)


# Create Objects
c = CEO()
m = Manager()
e = Employee()

# Store in List
employees = [c, m, e]

# Loop Through Objects
for emp in employees:
    show_details(emp)
```

```python
# Q4. YouTube, Netflix, Spotify, Podcast

class YouTube:
    def play(self, title):
        print(f"Playing YouTube video: {title}")

    def pause(self):
        print("YouTube paused")

    def stop(self):
        print("YouTube stopped")


class Netflix:
    def play(self, title):
        print(f"Playing Netflix show: {title}")

    def pause(self):
        print("Netflix paused")

    def stop(self):
        print("Netflix stopped")


class Spotify:
    def play(self, title):
        print(f"Playing Spotify song: {title}")

    def pause(self):
        print("Spotify paused")

    def stop(self):
        print("Spotify stopped")


class Podcast:
    def play(self, title):
        print(f"Playing Podcast: {title}")

    def pause(self):
        print("Podcast paused")

    def stop(self):
        print("Podcast stopped")


def stream(platform, title):
    platform.play(title)
    platform.pause()
    platform.stop()


stream(YouTube(), "Python Tutorial")
stream(Netflix(), "Money Heist")
stream(Spotify(), "Believer")
stream(Podcast(), "AI Talk")
```

```python
# Q5. PDF, Word, Excel, PowerPoint

class PDF:
    def open(self, filename):
        print(f"Opening PDF: {filename}")

    def save(self, filename):
        print(f"Saving PDF: {filename}")

    def close(self):
        print("PDF Closed")


class Word:
    def open(self, filename):
        print(f"Opening Word File: {filename}")

    def save(self, filename):
        print(f"Saving Word File: {filename}")

    def close(self):
        print("Word Closed")


class Excel:
    def open(self, filename):
        print(f"Opening Excel File: {filename}")

    def save(self, filename):
        print(f"Saving Excel File: {filename}")

    def close(self):
        print("Excel Closed")


class PowerPoint:
    def open(self, filename):
        print(f"Opening PPT: {filename}")

    def save(self, filename):
        print(f"Saving PPT: {filename}")

    def close(self):
        print("PPT Closed")


def handle_file(app, filename):
    app.open(filename)
    app.save(filename)
    app.close()


apps = [PDF(), Word(), Excel(), PowerPoint()]

for app in apps:
    handle_file(app, "demo.file")
    print("-" * 20)
```

```python
# Q6. BikeDelivery, CarDelivery, DroneDelivery, BoatDelivery

class BikeDelivery:
    def pickup(self, item):
        print(f"Bike picked up {item}")

    def deliver(self, item, address):
        print(f"Bike delivered {item} to {address}")

    def confirm(self):
        print("Bike Delivery Confirmed")


class CarDelivery:
    def pickup(self, item):
        print(f"Car picked up {item}")

    def deliver(self, item, address):
        print(f"Car delivered {item} to {address}")

    def confirm(self):
        print("Car Delivery Confirmed")


class DroneDelivery:
    def pickup(self, item):
        print(f"Drone picked up {item}")

    def deliver(self, item, address):
        print(f"Drone delivered {item} to {address}")

    def confirm(self):
        print("Drone Delivery Confirmed")


class BoatDelivery:
    def pickup(self, item):
        print(f"Boat picked up {item}")

    def deliver(self, item, address):
        print(f"Boat delivered {item} to {address}")

    def confirm(self):
        print("Boat Delivery Confirmed")


def process_delivery(vehicle, item, address):
    vehicle.pickup(item)
    vehicle.deliver(item, address)
    vehicle.confirm()


vehicles = [
    BikeDelivery(),
    CarDelivery(),
    DroneDelivery(),
    BoatDelivery()
]

for vehicle in vehicles:
    process_delivery(vehicle, "Laptop", "Hyderabad")
    print("-" * 20)
```

```python
# Q7. CreditCard, DebitCard, UPI, NetBanking (Duck Typing + Encapsulation)

class CreditCard:
    def __init__(self, balance):
        self.__balance = balance

    def validate(self, amount):
        return amount <= self.__balance

    def pay(self, amount):
        if self.validate(amount):
            self.__balance -= amount
            print(f"Credit Card Payment of ₹{amount} Successful")
        else:
            print("Insufficient Balance")

    def receipt(self):
        print("Receipt Generated")


class DebitCard:
    def __init__(self, balance):
        self.__balance = balance

    def validate(self, amount):
        return amount <= self.__balance

    def pay(self, amount):
        if self.validate(amount):
            self.__balance -= amount
            print(f"Debit Card Payment of ₹{amount} Successful")
        else:
            print("Insufficient Balance")

    def receipt(self):
        print("Receipt Generated")


class UPI:
    def __init__(self, balance):
        self.__balance = balance

    def validate(self, amount):
        return amount <= self.__balance

    def pay(self, amount):
        if self.validate(amount):
            self.__balance -= amount
            print(f"UPI Payment of ₹{amount} Successful")
        else:
            print("Insufficient Balance")

    def receipt(self):
        print("Receipt Generated")


class NetBanking:
    def __init__(self, balance):
        self.__balance = balance

    def validate(self, amount):
        return amount <= self.__balance

    def pay(self, amount):
        if self.validate(amount):
            self.__balance -= amount
            print(f"NetBanking Payment of ₹{amount} Successful")
        else:
            print("Insufficient Balance")

    def receipt(self):
        print("Receipt Generated")


def make_payment(method, amount):
    method.pay(amount)
    method.receipt()


payments = [
    CreditCard(5000),
    DebitCard(3000),
    UPI(2000),
    NetBanking(1000)
]

for p in payments:
    make_payment(p, 1500)
    print("-" * 30)
```

```python
# Q8. EmailAlert, SMSAlert, WhatsAppAlert, PushAlert

class EmailAlert:
    def __init__(self, recipient):
        self.__recipient = recipient
        self.__message_count = 0

    def send(self, message):
        self.__message_count += 1
        print(f"Email sent: {message}")

    def get_count(self):
        return self.__message_count

    @staticmethod
    def is_valid_message(msg):
        return msg != ""

    def show_stats(self):
        print("Recipient:", self.__recipient)
        print("Messages Sent:", self.__message_count)


class SMSAlert:
    def __init__(self, recipient):
        self.__recipient = recipient
        self.__message_count = 0

    def send(self, message):
        self.__message_count += 1
        print(f"SMS sent: {message}")

    def get_count(self):
        return self.__message_count

    @staticmethod
    def is_valid_message(msg):
        return msg != ""

    def show_stats(self):
        print("Recipient:", self.__recipient)
        print("Messages Sent:", self.__message_count)


class WhatsAppAlert:
    def __init__(self, recipient):
        self.__recipient = recipient
        self.__message_count = 0

    def send(self, message):
        self.__message_count += 1
        print(f"WhatsApp sent: {message}")

    def get_count(self):
        return self.__message_count

    @staticmethod
    def is_valid_message(msg):
        return msg != ""

    def show_stats(self):
        print("Recipient:", self.__recipient)
        print("Messages Sent:", self.__message_count)


class PushAlert:
    def __init__(self, recipient):
        self.__recipient = recipient
        self.__message_count = 0

    def send(self, message):
        self.__message_count += 1
        print(f"Push Notification: {message}")

    def get_count(self):
        return self.__message_count

    @staticmethod
    def is_valid_message(msg):
        return msg != ""

    def show_stats(self):
        print("Recipient:", self.__recipient)
        print("Messages Sent:", self.__message_count)


def alert_user(alert, message):
    if alert.is_valid_message(message):
        alert.send(message)


alerts = [
    EmailAlert("spandan@gmail.com"),
    SMSAlert("9876543210"),
    WhatsAppAlert("Spandan"),
    PushAlert("User")
]

for alert in alerts:
    alert_user(alert, "Welcome")
    alert_user(alert, "Offer Available")
    alert_user(alert, "Thank You")
    alert.show_stats()
    print("-" * 30)
```

```python
# Q9. Restaurant System (Duck Typing + Class Method + Static Method)

class Food:
    total_orders = 0

    def __init__(self, name, price):
        self.__name = name
        self.__price = price
        Food.total_orders += 1

    def prepare(self):
        print(f"Preparing {self.__name}")

    def serve(self):
        print(f"Serving {self.__name}")

    def bill(self):
        print(f"Bill Amount: ₹{self.__price}")

    @staticmethod
    def is_valid_price(price):
        return price > 0

    @classmethod
    def get_orders(cls):
        print("Total Orders:", cls.total_orders)


def process_order(food):
    food.prepare()
    food.serve()
    food.bill()


foods = [
    Food("Pizza", 250),
    Food("Burger", 150),
    Food("Sushi", 400),
    Food("Pasta", 200),
    Food("Salad", 100)
]

for item in foods:
    process_order(item)
    process_order(item)
    print("-" * 30)

Food.get_orders()
```

```python
# Q10. Bank Accounts System (Complete Polymorphism Project)

class Account:
    total_accounts = 0

    def __init__(self, owner, balance):
        self.__owner = owner
        self.__balance = balance
        Account.total_accounts += 1

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount

    def get_balance(self):
        return self.__balance

    def get_owner(self):
        return self.__owner

    @classmethod
    def get_total(cls):
        print("Total Accounts:", cls.total_accounts)


class SavingsAccount(Account):
    def calculate_interest(self):
        return self.get_balance() * 0.05


class CurrentAccount(Account):
    def calculate_interest(self):
        return self.get_balance() * 0.02


class LoanAccount(Account):
    def calculate_interest(self):
        return self.get_balance() * 0.10


class FixedDeposit(Account):
    def calculate_interest(self):
        return self.get_balance() * 0.08


class CryptoWallet(Account):
    def calculate_interest(self):
        return self.get_balance() * 0.15


def process_account(account, deposit_amt, withdraw_amt):
    account.deposit(deposit_amt)
    account.withdraw(withdraw_amt)

    print("Owner:", account.get_owner())
    print("Balance:", account.get_balance())
    print("Interest:", account.calculate_interest())
    print("-" * 30)


accounts = [
    SavingsAccount("Spandan", 10000),
    CurrentAccount("Rahul", 10000),
    LoanAccount("Amit", 10000),
    FixedDeposit("Riya", 10000),
    CryptoWallet("Raj", 10000)
]

for acc in accounts:
    process_account(acc, 2000, 1000)

Account.get_total()
```
